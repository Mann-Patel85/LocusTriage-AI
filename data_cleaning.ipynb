```python
import pandas as pd
import numpy as np
```

```python
print("📥 Fetching live data from NYC Open Data...")

# 1. Hit the official NYC 311 API (Built cleanly to avoid hidden characters)
base_url = "https://data.cityofnewyork.us/resource/erm2-nwe9.csv?$limit=200&"
# Expanded list of urban infrastructure issues for a more diverse dashboard
query = "$where=complaint_type in ('Street Condition', 'Street Light Condition', 'Water System', 'Sanitation Condition', 'Graffiti', 'Derelict Vehicles', 'Traffic Signal Condition', 'Sewer', 'Sidewalk Condition', 'Damaged Tree') AND latitude IS NOT NULL"
# Replace raw spaces with web-safe %20 characters
nyc_api_url = base_url + query.replace(" ", "%20")

# Load directly into a Pandas DataFrame from the internet!
df_raw = pd.read_csv(nyc_api_url)

print(f"✅ Successfully downloaded {len(df_raw)} records!")
```

```python
# 2. Format the columns to perfectly match LocusTriage AI
df_clean = pd.DataFrame()

df_clean['Issue_ID'] = "NYC-" + df_raw['unique_key'].astype(str)
df_clean['Latitude'] = df_raw['latitude']
df_clean['Longitude'] = df_raw['longitude']
df_clean['Category'] = df_raw['complaint_type']

# Simulate an Urgency Score (1-10) for the dashboard
df_clean['Urgency_Score'] = np.random.randint(3, 10, size=len(df_raw)) 
df_clean['Description'] = df_raw['descriptor']
```

```python
# 3. Export directly into your project's data folder
export_path = 'data/historical_issues.csv'
df_clean.to_csv(export_path, index=False)
```
